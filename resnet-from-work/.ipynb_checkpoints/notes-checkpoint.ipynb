## Create DataLoaders

Parse the directory structure and create (parallel) lists of **labels** and **filenames**. 

```python
train_path = 'C:\\Users\\jonathan.shuai\\Documents\\datasets\\aircrafts-3-4-2019\\Train'
valid_path = 'C:\\Users\\jonathan.shuai\\Documents\\datasets\\aircrafts-3-4-2019\\Valid'

# Get the paths to the raw image files and put them in an array
train_files = []
valid_files = []

# Iterate through data folder directory recursively
for root, directories, filenames in os.walk(train_path):
    # Only get images (in subfolders)
    if not root == train_path:
        for filename in filenames:
            train_files.append(os.path.join(root, filename))

            # Repeat for valid            
for root, directories, filenames in os.walk(valid_path):
    if not root == valid_path:
        for filename in filenames:
            valid_files.append(os.path.join(root, filename))


train_labels = [r.split('\\')[-2] for r in train_files]
label_encoder = LabelEncoder()
train_labels = label_encoder.fit_transform(train_labels)
valid_labels = label_encoder.transform([r.split('\\')[-2] for r in valid_files])

# Set the image size if AUTO
if IMAGE_SIZE == 'AUTO':
    IMAGE_SIZE = cv2.imread(train_files[0]).shape[:2]
    
if N_CLASSES == 'AUTO':
    N_CLASSES = label_encoder.classes_.shape[0]
    
print('Train size: {train_size}\nValid_size: {valid_size}\nImage size: {image_size}'.format(train_size=len(train_files),
                                                                                             valid_size=len(valid_files),
                                                                                             image_size=IMAGE_SIZE))
```

```python
transforms = [
                dataloader.RandomTranslate(),
                dataloader.RandomCropResize(),
                dataloader.Blur(),
                dataloader.ColorJitter(), 
                dataloader.RandomAffine(180, 0.2, [0.0, 0.20], 15),
             ]
transformer = dataloader.Transformer(transforms)

preprocessor = dataloader.PerImageStandardization() # Probably don't use this for now
```

```python
train_dataloader = dataloader.DataLoader(train_files, train_labels, 
                                         batch_size=BATCH_SIZE, 
                                         image_size=IMAGE_SIZE, 
                                         transformer=transformer,
#                                          preprocessor=preprocessor
                                        )

valid_dataloader = dataloader.DataLoader(valid_files, valid_labels, 
                                         batch_size=BATCH_SIZE, 
                                         image_size=IMAGE_SIZE, 
                                         transformer=transformer,
#                                          preprocessor=preprocessor
                                        )

batch_images, batch_labels = next(train_dataloader.get_data())

plt.figure(figsize=(12, 8))
for i, (image, label) in enumerate(zip(batch_images, label_encoder.inverse_transform(batch_labels))):
    plt.subplot(2, 4, i + 1)
    plt.imshow(image)
    plt.axis('off')
    plt.title(label)

plt.show()
```

## Creating a model (resnet?)

```python
# From TensorFlow tutorial
def variable_summaries(var):
    """Attach a lot of summaries to a Tensor (for TensorBoard visualization).
    """
    with tf.name_scope('summaries'):
        mean = tf.reduce_mean(var)
        tf.summary.scalar('mean', mean)
    with tf.name_scope('stddev'):
        stddev = tf.sqrt(tf.reduce_mean(tf.square(var - mean)))
    tf.summary.scalar('stddev', stddev)
    tf.summary.scalar('max', tf.reduce_max(var))
    tf.summary.scalar('min', tf.reduce_min(var))

def conv2d(x, W, stride, padding='SAME'):
    conv_layer = tf.nn.conv2d(x, W, strides=[1, stride, stride, 1], padding=padding)
    return conv_layer

def weight_variable(shape, name):
    initial = tf.truncated_normal(shape, stddev=0.01)
    return tf.Variable(initial, name=name)

def bias_variable(shape, name):
    initial = tf.constant(0.01, shape=shape)
    return tf.Variable(initial, name=name)

def basic_block(input_, in_channels, out_channels, training, downsample=False, filter_size=3, stride=1):
    identity = input_
    
    if downsample:
        W0 = weight_variable([1, 1, in_channels, out_channels], name="W0")
        identity = conv2d(identity, W0, stride)
        identity = tf.layers.batch_normalization(identity, training=training)
    

    W1 = weight_variable([filter_size, filter_size, in_channels, out_channels], name="W1")
    input_ = conv2d(input_, W1, stride)
    input_ = tf.layers.batch_normalization(input_, training=training)
    input_ = tf.nn.relu(input_)
    
    W2 = weight_variable([filter_size, filter_size, out_channels, out_channels], name="W2")
    input_ = conv2d(input_, W2, 1)
    input_ = tf.layers.batch_normalization(input_, training=training)
    input_ += identity
    
    return tf.nn.relu(input_)

def make_layer(input_, in_channels, out_channels, blocks, training, filter_size=3, stride=1):
    if stride != 1:
        input_ = basic_block(input_, in_channels, out_channels, training=training, 
                    downsample=True, filter_size=filter_size, stride=stride)
    else:
        input_ = basic_block(input_, in_channels, out_channels, training=training, 
                    downsample=False, filter_size=filter_size, stride=stride)
    
    for _ in range(blocks - 1):
        input_ = basic_block(input_, out_channels, out_channels, training=training,
                            downsample=False, filter_size=filter_size, stride=1)

    return input_

def resnet34(input_, training, n_classes):
    W1 = weight_variable([7, 7, 3, 64], name="W1")
    input_ = conv2d(input_, W1, 2)
    input_ = tf.nn.max_pool(input_, ksize=[1, 2, 2, 1], strides=[1, 1, 1, 1], padding='VALID')
    input_ = make_layer(input_, in_channels=64, out_channels=64, training=training, blocks=3, filter_size=3, stride=1)
    input_ = make_layer(input_, in_channels=64, out_channels=128, training=training, blocks=4, filter_size=3, stride=2)
    input_ = make_layer(input_, in_channels=128, out_channels=256, training=training, blocks=6, filter_size=3, stride=2)
    input_ = make_layer(input_, in_channels=256, out_channels=512, training=training, blocks=3, filter_size=3, stride=2)
    
    # Average pool reduces to Bx1x1x512
    input_ = tf.reduce_mean(input_, axis=[1, 2], keepdims=True)
    
    # Squeeze and apply fc
    input_ = tf.squeeze(input_, axis=[1, 2])
    input_ = tf.layers.dense(input_, n_classes)
    
    return input_
```

## Training

```python
training = True
# Create placeholders for labels and inputs
with tf.Graph().as_default() as graph:
    inputs = tf.placeholder(tf.float32, shape=[None, *IMAGE_SIZE, 3])
    labels = tf.placeholder(tf.int64, shape=[None,])
    
    output_ = resnet34(inputs, training, N_CLASSES)
        
    loss = tf.losses.sparse_softmax_cross_entropy(
        labels=labels, logits=output_
    )
    
    with tf.name_scope('accuracy'):
        prediction = tf.argmax(output_, 1)
        equality = tf.equal(prediction, labels)
        accuracy = tf.reduce_mean(tf.cast(equality, tf.float32))
        
    with tf.name_scope('train'):        
        # Create a scheduler to multiply learning rate by GAMMA every STEP_SIZE steps
        global_step = tf.Variable(0, trainable=False)
        scheduler = tf.train.exponential_decay(LEARNING_RATE, 
                                               global_step=global_step, 
                                               decay_steps=STEP_SIZE * train_dataloader.shape[0], 
                                               decay_rate=GAMMA, 
                                               staircase=True)
        
        # Create an optimizer (gradient descent) and have it minimize the loss
        optimizer = tf.train.GradientDescentOptimizer(scheduler)
#         optimizer = tf.train.AdamOptimizer(LEARNING_RATE)

        train_step = optimizer.minimize(loss, global_step=global_step)
    
    with tf.name_scope('logging'):
        variable_summaries(output_)
        tf.summary.scalar('current_loss', loss)
        tf.summary.histogram('prediction', prediction)
        summary = tf.summary.merge_all()

with tf.Session(graph=graph) as sess:
    if DEBUG:
        session = tf_debug.LocalCLIDebugWrapperSession(session)
    
    # Create a saver to make checkpoints and save final model
    saver = tf.train.Saver()
    
    # Initialize all weights: for the newly added retraining layer to random initial values.
    init = tf.global_variables_initializer()
    sess.run(init)

    # Create the log writers for the summary
    train_writer = tf.summary.FileWriter("{}/train".format(LOG_DIR), sess.graph)
    valid_writer = tf.summary.FileWriter("{}/valid".format(LOG_DIR), sess.graph)
    print('tensorboard --logdir={} --host=127.0.0.1'.format(LOG_DIR))

    for epoch in range(N_EPOCHS):
        print("Epoch {}".format(epoch))
        print("Training...")

        running_train_loss = 0
        running_train_acc = 0
        running_valid_loss = 0
        running_valid_acc = 0
        
        for batch_images, batch_labels in train_dataloader.get_data():
            _, train_loss, train_acc, train_summary = sess.run([train_step, loss, accuracy, summary], 
                                                               feed_dict={inputs: batch_images, labels: batch_labels})

            running_train_acc += train_acc * batch_images.shape[0]
            running_train_loss += train_loss

        running_train_acc /= train_dataloader.shape[0]

        for batch_images, batch_labels in valid_dataloader.get_data():
            valid_loss, valid_acc, valid_summary = sess.run([loss, accuracy, summary],
                                                           feed_dict={inputs: batch_images, labels: batch_labels})
            running_valid_acc += valid_acc * batch_images.shape[0]
            running_valid_loss += valid_loss

        running_valid_acc /= valid_dataloader.shape[0]

        # Print some diagnostics
        print("Train loss: {:.2f}\nTrain acc: {:.6f}".format(running_train_loss, 
                                                        running_train_acc))
        print("Valid loss: {:.2f}\nValid acc: {:.6f}".format(running_valid_loss, 
                                                                running_valid_acc))
        # Build on summary
        train_writer.add_summary(train_summary, epoch)
        valid_writer.add_summary(valid_summary, epoch)

        if (epoch % 10 == 0):
            save_path = saver.save(session, "./checkpoints/trained_model_{}.ckpt".format(epoch))
            print("Checkpoint: {}".format(save_path))
```
